```python
def get_movie_value(soup, field_name):
    
    '''Grab a value from Box Office Mojo HTML
    
    Takes a string attribute of a movie on the page and returns the string in
    the next sibling object (the value for that attribute) or None if nothing is found.
    '''
    
    obj = soup.find(text=re.compile(field_name))
    
    if not obj: 
        return None
    
    # this works for most of the values
    next_element = obj.findNext()
    
    if next_element:
        return next_element.text 
    else:
        return None
```

```python
def title_scrape(soup):
    
    try:    
        obj = soup.find(class_='title_wrapper').find('h1')
        text = obj.text.strip().split('\xa0')[0]
        return text
    except:
        return None
    
```

### Opening Weekend USA

```python
soup.find(id='titleDetails').find(text=re.compile('Opening Weekend')).findNext().previous_sibling
```

#### Function to extract opening weekend data from title page

```python
def opening_weekend(soup):
    
    try:
        obj = soup.find(id='titleDetails').find(text=re.compile('Opening Weekend'))
        text = obj.findNext().previous_sibling.strip()
    
        text = text.replace('\n', '').replace('$', '').replace(',', '')
        return int(text)
    except:
        return None
    
```

```python
opening_weekend(soup)
```

```python
def gross_rev(soup):
    
    try:
        obj = soup.find(id='titleDetails').find(text=re.compile('Gross USA'))
        text = obj.findNext().previous_sibling.previous_sibling.text.strip().split()[2]
    
        text = text.replace('$', '').replace(',', '')
        return int(text)
    except:
        return None
    
```

### Genres

```python
len(soup.find(id='titleStoryLine').find(text=re.compile('Genre')).findNext().findNext().text.strip())
```

#### Function to extract genre data from title page

```python
def genres(soup):
    
    try:
        obj = soup.find(id='titleStoryLine').find(text=re.compile('Genre'))
        genre_list = []
        str_check = bool(obj.findNext().text)
    
        while str_check:
            if len(obj.findNext().text.strip()) > 1:
                genre_list.append(obj.findNext().text.strip())
            obj = obj.findNext()
            str_check = bool(obj.findNext().text)
        return genre_list
    except:
        return None
    
    
```

```python
genres(soup)
```

```python
def get_movie_dict(link):
    
    url_base = 'https://www.imdb.com'
    
    url = url_base + link
    response = requests.get(url)
    soup = BeautifulSoup(response.text)

    
    headers = ['title', 'opn_wknd', 'gross', 'budget',
               'runtime', 'genres', 'date', 'mpaa' ]
    
    #Create movie dictionary and return
    movie_dict = dict(zip(headers, [title_scrape(soup),
                                opening_weekend(soup),
                                gross_rev(soup),
                                budget(soup), 
                                runtime(soup), 
                                genres(soup),
                                date_scrape(soup),
                                mpaa(soup)]))

    return movie_dict
```

```python
import time
import random

data_list = []
count = 0
for title in url_list:
    data_list.append(get_movie_dict(title))
    count += 1
    if count % 100 == 0:
        with open('IMDb_data_list.pickle', 'wb') as to_write:
            pickle.dump(data_list, to_write)
    time.sleep(.5+random.random())
```

### Getting cast/director/writer data

```python
url = 'https://www.imdb.com/title/tt0816692/'
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.text)
```

```python
soup.find('div', class_='credit_summary_item').find('a').text
```

```python
def director(soup):
    
    directors = []
    try:
        obj_list = soup.find_all('div', class_='credit_summary_item')[0].find_all('a')
        for director in obj_list:
            directors.append(director.text)
        return directors
    except:
        return None
```

```python
soup.find_all('div', class_='credit_summary_item')[1].find_all('a')
```

```python
def writers(soup):
    
    writers = []
    try:
        obj_list = soup.find_all('div', class_='credit_summary_item')[1].find_all('a')
        for writer in obj_list:
            writers.append(writer.text)
        return writers
    except:
        return None
```

```python
director(soup)
```

```python
writers(soup)
```

```python
soup.find('table', class_='cast_list').find_all('td', class_='primary_photo')
```

```python
def cast(soup):
    
    actors = []
    try:
        obj = soup.find('table', class_='cast_list').find_all('td', class_='primary_photo')
        for actor in obj:
            text = actor.findNext().findNext().findNext().text.split('\n')[1].strip()
            actors.append(text)
        return actors
    except:
        return None
```

```python
cast(soup)
```

```python
def get_personel_dict(link):
    
    url_base = 'https://www.imdb.com'
    
    url = url_base + link
    response = requests.get(url)
    soup = BeautifulSoup(response.text)

    
    headers = ['title', 'directors', 'writers', 'cast']
    
    #Create movie dictionary and return
    personel_dict = dict(zip(headers, [title_scrape(soup),
                                director(soup),
                                writers(soup),
                                cast(soup)]))

    return personel_dict
```

```python
crew_list = []

count = 0
for title in url_list:
    crew_list.append(get_personel_dict(title))
    count += 1
    if count % 100 == 0:
        with open('IMDb_crew_list.pickle', 'wb') as to_write:
            pickle.dump(crew_list, to_write)
    time.sleep(.5+random.random())
```
